```python
import os
import shutil, glob
import re
import csv
import matplotlib.pyplot as plt
import PIL
```

```python
file_path = "resistor_data"
resistor = './resistors/'
os.path.isdir(resistor)
#if os.path.isdir('file_path') is False:
#    os.mkdir('res_data')
```

```python
for i in range(len(os.listdir(file_path))):
    for file in os.listdir(file_path):
        file = "./resistor_data/"+file
        shutil.copy(file, resistor)
```

```python
files = []
file_num = '000000'
for i,file in enumerate(os.listdir(resistor), start=1):
    num = str(i)
    lidx = len(num)
    file_name = file_num[lidx:]+num
    old = resistor+file
    new = resistor+file_name+".jpg"
    os.rename(src=old, dst=new)
```

```python
label_file = []
for l in os.listdir(file_path):
    
    pattern = re.compile(r"([\(])|([0-9])|([\)])|(\.)|(jpg|jpeg|png|gif)")
    label = pattern.sub("", l)
    label_file.append(label)
```
